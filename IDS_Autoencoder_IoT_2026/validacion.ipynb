```python
# ==========================================================
# CELDA 1 - IMPORTAR LIBRERÍAS
# ==========================================================

import pandas as pd
import numpy as np
import os
```

```python
# ==========================================================
# CELDA 2 - CARGAR DATASETS
# ==========================================================

benign_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\Benign\benign_total.csv"

ddos_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\DDoS\ddos_total.csv"

bruteforce_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\BruteForce\bruteforce_total.csv"

# Cargar CSV
benign_df = pd.read_csv(benign_path)

ddos_df = pd.read_csv(ddos_path)

bruteforce_df = pd.read_csv(bruteforce_path)

print("Datasets cargados correctamente.")
```

```python
# ==========================================================
# CELDA 3 - VERIFICAR DIMENSIONES
# ==========================================================

print("Benign:", benign_df.shape)

print("DDoS:", ddos_df.shape)

print("BruteForce:", bruteforce_df.shape)
```

```python
# ==========================================================
# CELDA 4 - VERIFICAR COLUMNAS
# ==========================================================

benign_cols = set(benign_df.columns)

ddos_cols = set(ddos_df.columns)

bruteforce_cols = set(bruteforce_df.columns)

print("Columnas faltantes en DDoS:")
print(benign_cols - ddos_cols)

print("\nColumnas extra en DDoS:")
print(ddos_cols - benign_cols)

print("\nColumnas faltantes en BruteForce:")
print(benign_cols - bruteforce_cols)

print("\nColumnas extra en BruteForce:")
print(bruteforce_cols - benign_cols)
```

```python
# ==========================================================
# CELDA 5 - VERIFICAR TIPOS DE DATOS
# ==========================================================

print("\n================ BENIGN ================\n")
print(benign_df.dtypes)

print("\n================ DDoS ================\n")
print(ddos_df.dtypes)

print("\n================ BRUTE FORCE ================\n")
print(bruteforce_df.dtypes)
```

```python
# ==========================================================
# CELDA 6 - BUSCAR VALORES NULOS (NaNs)
# ==========================================================

datasets = {
    "Benign": benign_df,
    "DDoS": ddos_df,
    "BruteForce": bruteforce_df
}

for name, dataset in datasets.items():

    print(f"\n=========== {name} ===========")

    nulls = dataset.isnull().sum()

    nulls = nulls[nulls > 0]

    if len(nulls) == 0:
        print("No hay valores nulos.")

    else:
        print(nulls.sort_values(ascending=False))
```

```python
# ==========================================================
# CELDA 7 - BUSCAR VALORES INFINITOS
# ==========================================================

for name, dataset in datasets.items():

    numeric_df = dataset.select_dtypes(include=[np.number])

    inf_count = np.isinf(numeric_df).sum().sum()

    print(f"{name}: {inf_count} valores infinitos")
```

```python
# ==========================================================
# CELDA 8 - VERIFICAR LABELS
# ==========================================================

print("\nBENIGN")
print(benign_df["Label"].value_counts())

print("\nDDoS")
print(ddos_df["Label"].value_counts())

print("\nBRUTE FORCE")
print(bruteforce_df["Label"].value_counts())
```

```python
# ==========================================================
# CELDA 9 - VERIFICAR DUPLICADOS
# ==========================================================

print("Duplicados Benign:", benign_df.duplicated().sum())

print("Duplicados DDoS:", ddos_df.duplicated().sum())

print("Duplicados BruteForce:", bruteforce_df.duplicated().sum())
```

```python
# ==========================================================
# CELDA 10 - COLUMNAS COMPLETAMENTE VACÍAS
# ==========================================================

for name, dataset in datasets.items():

    empty_cols = dataset.columns[dataset.isnull().all()]

    print(f"\n{name}")

    if len(empty_cols) == 0:
        print("No hay columnas completamente vacías.")

    else:
        print(list(empty_cols))
```

```python
# ==========================================================
# CELDA 11 - MEMORIA UTILIZADA
# ==========================================================

for name, dataset in datasets.items():

    memory = dataset.memory_usage(deep=True).sum() / 1024**2

    print(f"{name}: {memory:.2f} MB")
```
