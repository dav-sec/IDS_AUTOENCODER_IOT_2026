```python
# ==========================================================
# CELDA 1 - IMPORTACIÓN DE LIBRERÍAS
# ==========================================================

import pandas as pd
import numpy as np
```

```python
# ==========================================================
# CELDA 2 - CARGA DE DATASETS
# ==========================================================

# Dataset benigno
benign_df = pd.read_csv(
    r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\Benign\benign_total.csv"
)

# Dataset DDoS
ddos_df = pd.read_csv(
    r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\DDoS\ddos_total.csv"
)

# Dataset Brute Force
bruteforce_df = pd.read_csv(
    r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\BruteForce\bruteforce_total.csv"
)

# Mostrar dimensiones
print("Benign:", benign_df.shape)
print("DDoS:", ddos_df.shape)
print("BruteForce:", bruteforce_df.shape)
```

```python
# ==========================================================
# CELDA 3 - UNIFICACIÓN DE DATASETS
# ==========================================================

# Concatenar datasets
df = pd.concat(
    [benign_df, ddos_df, bruteforce_df],
    ignore_index=True
)

# Mostrar dimensiones
print("Dataset unificado:")
print(df.shape)

# Mostrar primeras filas
df.head()
```

```python
# ==========================================================
# CELDA 4 - VISUALIZAR COLUMNAS
# ==========================================================

for col in df.columns:
    print(col)
```

```python
# ==========================================================
# CELDA 5 - ELIMINAR FEATURES IRRELEVANTES
# ==========================================================

# Keywords problemáticas
columns_to_remove_keywords = [
    "ip",
    "port",
    "mac",
    "timestamp",
    "first_seen",
    "last_seen",
    "flow_id",
    "tunnel",
    "vlan",
    "fingerprint",
    "server_name",
    "user_agent",
    "dns",
    "oui"
]

columns_to_drop = []

# Buscar columnas automáticamente
for col in df.columns:

    col_lower = col.lower()

    for keyword in columns_to_remove_keywords:

        if keyword in col_lower:

            columns_to_drop.append(col)

            break

# Eliminar columnas tipo object excepto Label
object_columns = df.select_dtypes(include=['object']).columns

for col in object_columns:

    if col != "Label" and col not in columns_to_drop:

        columns_to_drop.append(col)

# Eliminar duplicados en lista
columns_to_drop = list(set(columns_to_drop))

# Mostrar columnas eliminadas
print("Columnas eliminadas:\n")

for col in columns_to_drop:
    print(col)

# Eliminar columnas
df = df.drop(columns=columns_to_drop, errors='ignore')

# Mostrar dimensiones
print("\nNuevas dimensiones:")
print(df.shape)
```

```python
# ==========================================================
# CELDA 6 - DETECCIÓN DE VALORES NULOS
# ==========================================================

# Contar NaNs
null_values = df.isnull().sum()

# Mostrar solo columnas con NaNs
null_values = null_values[null_values > 0]

print(null_values)
```

```python
# ==========================================================
# CELDA 7 - IMPUTACIÓN DE VALORES NULOS
# ==========================================================

# Seleccionar columnas numéricas
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Imputar NaNs con media
for col in numeric_columns:

    mean_value = df[col].mean()

    df[col] = df[col].fillna(mean_value)

# Verificar si aún quedan NaNs
remaining_nulls = df.isnull().sum().sum()

print("Valores nulos restantes:", remaining_nulls)
```

```python
# ==========================================================
# CELDA 8 - ELIMINAR DUPLICADOS
# ==========================================================

print("Antes:", df.shape)

# Eliminar duplicados
df = df.drop_duplicates().reset_index(drop=True)

print("Después:", df.shape)
```

```python
# ==========================================================
# CELDA 9 - ELIMINAR COLUMNAS CASI VACÍAS
# ==========================================================

threshold = 0.99

columns_to_remove = []

# Solo columnas numéricas
numeric_columns = df.select_dtypes(include=[np.number]).columns

for col in numeric_columns:

    if col == "Label":
        continue

    zero_ratio = (df[col] == 0).mean()

    if zero_ratio > threshold:

        columns_to_remove.append(col)

print("Columnas eliminadas:\n")

for col in columns_to_remove:
    print(col)

# Eliminar columnas
df = df.drop(columns=columns_to_remove)

print("\nDimensiones finales:")
print(df.shape)
```

```python
# ==========================================================
# CELDA 10 - EXPORTAR DATASET LIMPIO
# ==========================================================

output_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\dataset_limpio.csv"

# Guardar dataset limpio
df.to_csv(output_path, index=False)

print("Dataset limpio exportado correctamente.")
print(f"Ruta: {output_path}")
```
