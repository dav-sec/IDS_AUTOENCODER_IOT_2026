```python
# ==========================================================
# IMPORTACIÓN DE LIBRERÍAS
# ==========================================================

from nfstream import NFStreamer
import pandas as pd
import os
```

```python
# ==========================================================
# EXTRACCIÓN DE FLOWS - TRÁFICO BENIGNO
# ==========================================================

# Carpeta de entrada
input_folder = r"E:\Proyectos_GitHub\Procesamiento\Dataset\Benign_Final"

# Carpeta de salida
output_folder = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\Benign"

# Lista de archivos PCAP benignos
pcap_files = [
    "BenignTraffic.pcap",
    "BenignTraffic1.pcap",
    "BenignTraffic2.pcap",
    "BenignTraffic3.pcap"
]

# Procesar cada archivo
for pcap_file in pcap_files:

    print(f"\nProcesando: {pcap_file}")

    # Ruta completa del PCAP
    pcap_path = os.path.join(input_folder, pcap_file)

    # Crear streamer NFStream
    streamer = NFStreamer(
        source=pcap_path,
        statistical_analysis=True
    )

    # Convertir flows a DataFrame
    flows_df = streamer.to_pandas()

    # Nombre CSV
    csv_name = pcap_file.replace(".pcap", ".csv")

    # Ruta de salida
    output_csv = os.path.join(output_folder, csv_name)

    # Guardar CSV
    flows_df.to_csv(output_csv, index=False)

    print(f"CSV guardado en: {output_csv}")
    print(f"Dimensiones: {flows_df.shape}")

print("\nTodos los archivos BENIGN fueron procesados.")
```

```python
# ==========================================================
# EXTRACCIÓN DE FLOWS - ATAQUE DDoS UDP FLOOD
# ==========================================================

# Carpeta de entrada
input_folder = r"E:\Proyectos_GitHub\Procesamiento\Dataset\DDoS-UDP_Flood"

# Carpeta de salida
output_folder = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\DDoS"

# Generar automáticamente nombres:
# DDoS-UDP_Flood.pcap
# DDoS-UDP_Flood1.pcap
# ...
# DDoS-UDP_Flood20.pcap

pcap_files = ["DDoS-UDP_Flood.pcap"] + [
    f"DDoS-UDP_Flood{i}.pcap" for i in range(1, 21)
]

# Procesar archivos
for pcap_file in pcap_files:

    print(f"\nProcesando: {pcap_file}")

    # Ruta completa
    pcap_path = os.path.join(input_folder, pcap_file)

    # Crear streamer
    streamer = NFStreamer(
        source=pcap_path,
        statistical_analysis=True
    )

    # Convertir a DataFrame
    flows_df = streamer.to_pandas()

    # Nombre CSV
    csv_name = pcap_file.replace(".pcap", ".csv")

    # Ruta salida
    output_csv = os.path.join(output_folder, csv_name)

    # Exportar CSV
    flows_df.to_csv(output_csv, index=False)

    print(f"CSV guardado en: {output_csv}")
    print(f"Dimensiones: {flows_df.shape}")

print("\nTodos los archivos DDoS fueron procesados.")
```
