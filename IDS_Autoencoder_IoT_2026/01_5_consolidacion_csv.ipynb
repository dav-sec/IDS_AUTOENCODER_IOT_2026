```python
# ==========================================================
# CELDA 1 - IMPORTACIÓN DE LIBRERÍAS
# ==========================================================

import pandas as pd
import os
```

```python
# ==========================================================
# CELDA 2 - CONSOLIDAR CSV BENIGNOS
# ==========================================================

# Carpeta donde están los CSV benignos
folder_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\Benign"

# Lista de DataFrames
dfs = []

# Recorrer archivos
for file in os.listdir(folder_path):

    if file.endswith(".csv"):

        file_path = os.path.join(folder_path, file)

        print(f"Cargando: {file}")

        df = pd.read_csv(file_path)

        # Etiqueta
        df["Label"] = "Benign"

        dfs.append(df)

# Unificar
benign_df = pd.concat(dfs, ignore_index=True)

# Guardar CSV final
output_path = os.path.join(folder_path, "benign_total.csv")

benign_df.to_csv(output_path, index=False)

print("\nCSV benign_total.csv generado correctamente.")
print(benign_df.shape)
```

```python
# ==========================================================
# CELDA 3 - CONSOLIDAR CSV DDoS
# ==========================================================

folder_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\DDoS"

dfs = []

for file in os.listdir(folder_path):

    if file.endswith(".csv"):

        file_path = os.path.join(folder_path, file)

        print(f"Cargando: {file}")

        df = pd.read_csv(file_path)

        df["Label"] = "DDoS"

        dfs.append(df)

ddos_df = pd.concat(dfs, ignore_index=True)

output_path = os.path.join(folder_path, "ddos_total.csv")

ddos_df.to_csv(output_path, index=False)

print("\nCSV ddos_total.csv generado correctamente.")
print(ddos_df.shape)
```

```python
# ==========================================================
# CELDA 4 - CONSOLIDAR CSV BRUTE FORCE
# ==========================================================

folder_path = r"E:\Proyectos_GitHub\Procesamiento\CSV_Output\BruteForce"

output_filename = "bruteforce_total.csv"

dfs = []

for file in os.listdir(folder_path):

    # Evitar leer el archivo consolidado
    if file.endswith(".csv") and file != output_filename:

        file_path = os.path.join(folder_path, file)

        print(f"Cargando: {file}")

        df = pd.read_csv(file_path)

        # Agregar etiqueta
        df["Label"] = "BruteForce"

        dfs.append(df)

# Concatenar todos los DataFrames
bruteforce_df = pd.concat(dfs, ignore_index=True)

# Ruta de salida
output_path = os.path.join(folder_path, output_filename)

# Guardar CSV final
bruteforce_df.to_csv(output_path, index=False)

print("\nCSV bruteforce_total.csv generado correctamente.")
print(bruteforce_df.shape)
```
